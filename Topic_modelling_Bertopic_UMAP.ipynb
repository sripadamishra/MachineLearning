```python
from sklearn.datasets import fetch_20newsgroups
docs = fetch_20newsgroups(subset='all',  remove=('headers', 'footers', 'quotes'))['data']
```

Text preprocessing is not shown here..you can do stop word removal, lemmatization...

We shall interest here the process of UMAP. BERTopic model by default produces different results each time because of stochasticity inherited from UMAP.

To get reproducible topics, we should pass random_state parameter in the UMAP method.
n_neighbour=15 indicates local neighbourhood size for UMAP is 15. This parameter controls the local versas global structure in data.

1. A low value forces UMAP to focus on local structure, might loose insights of the big picture.
2. A high value is to look the broader neighbourhood and may loose details on the local structure.
3. default value is 15.

n_components=5 indicates the target dimension from UMAP is 5. This is the dimension of data that will be passed into the clustering model.

min_dist controls how UMAP can pack all points together. It is the minimum distance between points in low dimensional space.
1. Small value of min_dist indicates clumpier embeddings, good for clustering. Since our goal here is to build clustering models, so min_dist=0 is set here.
2. Large value of min_dist prevent UMAP from packings together and preserve the broad structure of data.

metric='cosine' indicates that cosine isused to measure the distance.

random_state sets the random seed to make the UMAP results reproducible.


#Initiate UMAP

```python
from umap import UMAP
from bertopic import BERTopic
umap_model = UMAP(n_neighbors=15,
                  n_components=5,
                  min_dist=0.0,
                  metric='cosine',
                  random_state=100)

#initiate BERTopic with UMAP

topic_model = BERTopic(umap_model=umap_model,language="english",calculate_probabilities=True)

topics,probabilities = topic_model.fit_transform(docs)
```
